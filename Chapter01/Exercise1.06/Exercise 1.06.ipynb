```python
import tensorflow as tf
w=tf.Variable(0.0)
b=tf.Variable(0.0)
```

```python
def regression(x):
    model=w*x+b
    return model
```

```python
x=[1,2,3,4]
y=[0,-1,-2,-3]
```

```python
loss=lambda:abs(regression(x)-y)
```

```python
optimizer=tf.optimizers.Adam(.01)
```

```python
for i in range(1000):
    optimizer.minimize(loss,[w,b])
```
